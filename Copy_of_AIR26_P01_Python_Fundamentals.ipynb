# 7. Loops

Loops are used to repeat a task.

Python commonly uses:

- `for` loop
- `while` loop

Loops are useful in robotics because sensors keep producing data repeatedly.

````python
# for loop example

sensor_readings = [0.9, 0.7, 0.4, 1.2]

for distance in sensor_readings:
    print("Obstacle distance:", distance, "m")
````

````python
# Robot decision for each sensor reading

sensor_readings = [0.9, 0.7, 0.4, 1.2, 0.3]

for distance in sensor_readings:
    if distance < 0.5:
        print(distance, "m → Stop")
    else:
        print(distance, "m → Move")
````

````python
# range() is used to generate a sequence of numbers.

for step in range(5):
    print("Robot step:", step)
````

## More about `range()`

The `range()` function can be used in three common ways:

```python
range(stop)
range(start, stop)
range(start, stop, step)
```

Important point:

```text
The stop value is not included.
```

````python
# range(start, stop)
# This prints robot step numbers from 1 to 5.
# Note: The stop value is not included.

for step in range(1, 6):
    print("Robot step:", step)
````

````python
# range(start, stop, step)
# This prints even numbers from 0 to 8.

for value in range(0, 10, 2):
    print(value)
````

## `break` and `continue`

- `break` stops the loop immediately.
- `continue` skips the current iteration and moves to the next one.

````python
# Stop checking once a dangerous obstacle is found.

sensor_readings = [1.0, 0.8, 0.6, 0.3, 0.9]

for distance in sensor_readings:
    if distance < 0.5:
        print("Dangerous obstacle found:", distance, "m")
        break
    print("Safe distance:", distance, "m")
````

# 8. Functions

A function is a reusable block of code.

Functions help us avoid repeating the same code again and again.

Basic structure:

```python
def function_name(parameters):
    statements
    return result
```

## Parameter vs Argument

A **parameter** is the variable written inside the function definition.

An **argument** is the actual value passed when calling the function.

Example:

```python
def decide_robot_action(distance):
    ...
```

Here, `distance` is a **parameter**.

```python
decide_robot_action(0.3)
```

Here, `0.3` is an **argument**.

````python
# Simple function example

def show_robot_name(name):
    print("Robot name is:", name)

show_robot_name("AIR Robot")
````

````python
# Function to decide robot action

def decide_robot_action(distance):
    if distance < 0.5:
        return "Stop"
    else:
        return "Move forward"

print(decide_robot_action(0.3))
print(decide_robot_action(0.8))
````

````python
# Function with two inputs

def check_robot_status(battery_level, obstacle_distance):
    if battery_level < 20:
        return "Charge battery"
    elif obstacle_distance < 0.5:
        return "Stop due to obstacle"
    else:
        return "Ready to move"

print(check_robot_status(80, 1.0))
print(check_robot_status(10, 1.0))
print(check_robot_status(80, 0.3))
````

## Optional: Recursion — For Curious Students

This topic is not required for the first Machine Learning sessions.

Read this section only after you are comfortable with functions.

Recursion means a function calls itself.

Beginners do not need to master recursion immediately, but it is useful to understand the idea.

Example: factorial

```text
factorial(5) = 5 × 4 × 3 × 2 × 1 = 120
```

````python
# Optional: Factorial using recursion

def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n - 1)

print(factorial(5))
````

# 14. Common Errors and Debugging

Errors are normal in programming. A good programmer learns how to read and fix errors.

Common Python errors:

| Error | Common Reason |
|---|---|
| `SyntaxError` | Python grammar mistake |
| `NameError` | Variable name not defined |
| `TypeError` | Wrong data type used |
| `IndentationError` | Incorrect spacing |
| `ZeroDivisionError` | Dividing by zero |

Below we show corrected examples. The incorrect examples are shown as comments so that the notebook runs without stopping.

````python
# NameError example
# Incorrect:
# print(robot_speed)

# Correct:
robot_speed = 0.5
print(robot_speed)
````

````python
# ZeroDivisionError example
# Incorrect:
# speed = distance / 0

# Correct:
distance = 10
time = 5
speed = distance / time
print(speed)
````

## Handling Errors Using `try-except`

Sometimes errors may occur during program execution.

Instead of stopping the whole program, we can handle errors using `try-except`.

This is useful in robotics because a robot program should handle unexpected situations safely.

````python
# Simple try-except example

distance = 10
time = 0

try:
    speed = distance / time
    print("Speed:", speed)
except ZeroDivisionError:
    print("Error: Time cannot be zero. Speed cannot be calculated.")
````

````python
# Handling wrong type conversion

value = "robot"

try:
    number = int(value)
    print(number)
except ValueError:
    print("Error: This text cannot be converted into a number.")
````

# 15. Mini Lab: Simple Robot Decision System

In this mini lab, we will combine:

- variables,
- lists,
- loops,
- conditions,
- functions,
- dictionaries.

Problem:

A robot receives obstacle-distance readings.  
For each reading:

- if distance is less than 0.5 m, the robot should stop,
- otherwise, the robot should move forward.

````python
# Mini Lab: Rule-based robot decision system

def decide_action(distance):
    if distance < 0.5:
        return "Stop"
    else:
        return "Move forward"

robot_status = {
    "name": "AIR Robot",
    "battery": 78,
    "mode": "autonomous"
}

sensor_readings = [1.2, 0.9, 0.45, 0.8, 0.3]

print("Robot:", robot_status["name"])
print("Mode:", robot_status["mode"])
print("Battery:", robot_status["battery"], "%")
print("-" * 40)

for reading in sensor_readings:
    action = decide_action(reading)
    print(f"Distance: {reading} m → Action: {action}")
````
